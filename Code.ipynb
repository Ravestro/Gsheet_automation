```python
from selenium import webdriver
import pygsheets

driver = webdriver.Edge()
website = 'https://www.imdb.com/chart/top/?sort=user_rating%2Cdesc'
driver.get(website)
output = driver.find_elements("xpath",'//h3[@class="ipc-title__text"]')
span_output = driver.find_elements("xpath",'//span[@class="sc-b189961a-8 kLaxqf cli-title-metadata-item"]')
rating_output =  driver.find_elements("xpath",'//span[@aria-label]')
```

```python
import pygsheets
import pandas as pd

url = 'https://docs.google.com/spreadsheets/d/1I9SODidq-0WZqIxJ0-YzS_tbcd5GMpADjo3H_h34Nl4/edit#gid=0'
gc = pygsheets.authorize(service_file = "C:\Become_Datascientist\Automation\gsheet-automation-423205-419bc18c169f.json")
sh = gc.open_by_url(url)
wks = sh.sheet1

id = [i for i in range(1,251)]
store = []
span_store = []
rating_store = []

for i in output[1:-13]:
    x = i.text
    x = x.split('.',1)
    store.append(x[1])

for j in range(0,len(span_output)-2,3):
    span_store.append([span_output[j].text,span_output[j+1].text,span_output[j+2].text])

for i in rating_output[1:-2:1]:
    x = i.text
    x = x.split('(')
    rating_store.append(x[0])

data = {
    'id': id,
    'Movie': store,
    'Time': [i[0] for i in span_store] ,
    'Rating': rating_store,
    'Year': [i[1] for i in span_store] 
    
}

df = pd.DataFrame(data)

wks.clear()

wks.set_dataframe(df, start='A1')

driver.quit()






```
