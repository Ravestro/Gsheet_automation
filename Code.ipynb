```python
import pygsheets

url = 'https://docs.google.com/spreadsheets/d/1I9SODidq-0WZqIxJ0-YzS_tbcd5GMpADjo3H_h34Nl4/edit#gid=0'
gc = pygsheets.authorize(service_file = 'gsheet-automation-423205-419bc18c169f.json')
sh = gc.open_by_url(url)
wks = sh.sheet1

id = [[i] for i in range(1,251)]
store = []
span_store = []
rating_store = []

for i in output[1:-13]:
    x = i.text
    x = x.split('.',1)
    store.append([x[1]])

for j in range(1,len(span_output)-2,3):
    span_store.append([span_output[j].text,span_output[j+1].text,span_output[j+2].text])

for i in rating_output[1:-2:1]:
    x = i.text
    x = x.split('(')
    rating_store.append([x[0]])

wks.clear()
wks.update_values('A2',id)
wks.update_values('B2',store)
wks.update_values('C2',span_store)
wks.update_values('D2',rating_store)
wks.update_values('A1',[['id','Movie','Time','Rating','Year']])




```

```python
# Create a chart
wks.add_chart(('A2', 'A7'), [('B1', 'B6')], 'Age Chart') 
```
